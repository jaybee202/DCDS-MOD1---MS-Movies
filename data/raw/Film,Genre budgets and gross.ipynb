```python
import pandas as pd
moviebudgets = pd.read_csv('tn.movie_budgets.csv.gz')
titlebasics = pd.read_csv('imdb.title.basics.csv.gz')
titleakas = pd.read_csv('imdb.title.akas.csv.gz')
titleprincipals = pd.read_csv('imdb.title.principals.csv.gz')
titlecrew = pd.read_csv('imdb.title.crew.csv.gz')
```

```python
mb = moviebudgets
tb = titlebasics
ak = titleakas
tc = titlecrew
```

```python
mb = mb.sort_values(['movie'], ascending=True)
mb.set_index('movie', inplace=True)
mb
```

```python
tb = tb.sort_values(['primary_title'], ascending=True)
tb.set_index('primary_title', inplace=True)
tb
```

```python
df = mb.join(tb, how='inner')
```
